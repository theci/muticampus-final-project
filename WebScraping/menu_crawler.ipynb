```python
import requests
import re
import warnings
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium import webdriver 

driver = webdriver.Chrome('C:/chromedriver_win32/chromedriver.exe')



# 불필요한 경고 출력을 방지합니다.
warnings.filterwarnings('ignore')
```

### 더본코리아 대표브랜드 음식점별 메뉴 정보 수집 크롤러

```python
def menu_crawler() :
    # -------------------------------------------------------------------------
    # -------------------------------------------------------------------------
    # 아래부터 원근 크롤러
    # -------------------------------------------------------------------------
    # -------------------------------------------------------------------------
    
    # 고투웍 크롤링
    url = "https://www.theborn.co.kr/theborn_brand/%ea%b3%a0%ed%88%ac%ec%9b%8d/#menu_list"
    html = requests.get(url, verify = False).text
    soup = bs(html,"lxml")
    menu_html = soup.select("div.menu_list > div.item > div.menu_info > p:nth-child(3)")

    menu_text = []
    for i in menu_html:
        text = i.get_text()
        menu_text.append(text)
    menu_merge = ",".join(menu_text)
    menu_merge_replace = menu_merge.replace(" ","").replace("\n","").replace("*메인메뉴는시기에따라변경될수있습니다.","")

    # 메뉴명
    menu_name = menu_merge_replace.split(",")

    # 가게이름 리스트 생성
    restaurant_name = []
    restaurant_category = []

    for num in range(len(menu_name)):
        restaurant_name.append("고투웍")
        restaurant_category.append("중식")
    
    gotowok_df = pd.DataFrame({"restaurant_name":restaurant_name,"menu_name":menu_name})

    # -------------------------------------------------------------------------
    # 빽보이 크롤링
    url = "https://www.theborn.co.kr/theborn_brand/%eb%b9%bd%eb%b3%b4%ec%9d%b4%ed%94%bc%ec%9e%90/#menu_list"
    html = requests.get(url, verify = False).text
    soup = bs(html,"lxml")
    menu_html = soup.select("div.item > div.menu_info > div > p:nth-child(1)")
    menu_info_html = soup.select("div.item > div.menu_info > p:nth-child(3)")

    menu_text = []
    for i in menu_html:
        text = i.get_text()
        menu_text.append(text)
    menu_merge = ",".join(menu_text)
    menu_merge_replace = menu_merge.replace(" ","").replace("\n","").replace("*메인메뉴는시기에따라변경될수있습니다.","")

    # 메뉴설명
    menu_info = []
    for i in menu_info_html:
        text = i.get_text()
        menu_info.append(text.replace("\n",""))

    # 메뉴명
    menu_name = menu_merge_replace.split(",")

    # 가게이름 리스트 생성
    restaurant_name = []
    restaurant_category = []

    for num in range(len(menu_name)):
        restaurant_name.append("빽보이")
        restaurant_category.append("양식")
    
    paikboy_df = pd.DataFrame({"restaurant_name":restaurant_name,"menu_name":menu_name})


    # -------------------------------------------------------------------------
    # 연돈볼카츠 크롤링
    url = "https://www.theborn.co.kr/theborn_brand/%ec%97%b0%eb%8f%88%eb%b3%bc%ec%b9%b4%ec%b8%a0/#menu_list"
    html = requests.get(url, verify = False).text
    soup = bs(html,"lxml")
    menu_html = soup.select("div.item > div.menu_info > div > p:nth-child(1)")
    menu_info_html = soup.select("div.item > div.menu_info > p:nth-child(3)")

    menu_text = []
    for i in menu_html:
        text = i.get_text()
        menu_text.append(text)
    menu_merge = ",".join(menu_text)
    menu_merge_replace = menu_merge.replace(" ","").replace("\n","").replace("*메인메뉴는시기에따라변경될수있습니다.","")

    # 메뉴설명
    menu_info = []
    for i in menu_info_html:
        text = i.get_text()
        menu_info.append(text.replace("\n",""))

    # 메뉴명
    menu_name = menu_merge_replace.split(",")

    # 가게이름 리스트 생성
    restaurant_name = []
    restaurant_category = []

    for num in range(len(menu_name)):
        restaurant_name.append("연돈볼카츠")
        restaurant_category.append("일식")
    
    yeondon_df = pd.DataFrame({"restaurant_name":restaurant_name,"menu_name":menu_name})

    # -------------------------------------------------------------------------
    # 리춘시장 데이터프레임
    restaurant_name = []
    restaurant_category = []
    menu_name = ["마라샹궈","마라탕","마라편","마라감튀","마라바지락볶음","마라볶음밥",\
        "사막깐풍기","향라고추윙봉","멘바샥","뚝배기마파두부","고추듬뿍유린기","뚝배기어향가지","토마토계란볶음"\
            "매운우삼겹건두부볶음","즈란닭꼬치","칠리새우","마요새우","중화바지락볶음","매운바지락볶음","새우완탕"]
    
    for i in range(len(menu_name)):
        restaurant_name.append("리춘시장")
        restaurant_category.append("중식")

    licun_df = pd.DataFrame({"restaurant_name":restaurant_name,"menu_name":menu_name})
    
    # -------------------------------------------------------------------------
    # 인생설렁탕 데이터프레임
    restaurant_name = []
    restaurant_category = []
    menu_name = ["인생설렁탕","시래기된장설렁탕","들깨해장탕","곱빼기설렁탕","스지설렁탕","옛날불고기","양곱창전골","인생수육한판","국물불스지","목포식수육무침","인생물냉면","인생비빔냉면"]

    for i in range(len(menu_name)):
        restaurant_name.append("인생설렁탕")
        restaurant_category.append("한식")

    insaeng_df = pd.DataFrame({"restaurant_name":restaurant_name,"menu_name":menu_name})

    # -------------------------------------------------------------------------
    # 원근 데이터프레임 합치기
    dataframe_wonkkkn = gotowok_df.append(paikboy_df).append(yeondon_df).append(licun_df).append(insaeng_df)
    dataframe_wonkkkn.reset_index(drop = True, inplace = True)

    # -------------------------------------------------------------------------
    # -------------------------------------------------------------------------
    # 아래부터 창훈 크롤러
    # -------------------------------------------------------------------------
    # -------------------------------------------------------------------------

    # 1번째 식당
    url = 'https://www.theborn.co.kr/theborn_brand/%ec%84%b1%ec%84%b1%ec%8b%9d%eb%8b%b9/' # 성성식당
    driver.get(url) 
    html = driver.page_source
    soup = bs(html, 'html.parser') 



    a = soup.find_all("div", {"class":"name"}) #메뉴 이름 위치 찾기

    food_list = [] #메뉴 이름을 담을 리스트
    for i in range(len(a)): #for문을 돌면서 식당 추가
        name = a[i].get_text().strip()
        food_list.append(name)

    df1 = pd.DataFrame(food_list, columns = ['menu']) # 칼럼명이 menu로 데이터프레임 열 하나 생성
    df1.insert(0,'restaurant_name', '성성식당') # 식당명과 카테고리 열 생성




    # 2번째 식당
    url = 'https://www.paiks-pan.com/menu/' #백철판
    driver.get(url) 
    html = driver.page_source
    soup = bs(html, 'html.parser') 

    # 메뉴 구역이 나눠져있어서 따로 코드 작성
    food_list1 = []
    b = soup.find("section", {"class":"section menu_01"}).find_all("h3")

    for i in range(len(b)):
        b1 = b[i].get_text()
        food_list1.append(b1)

    food_list2 = []
    b2 = soup.find("section", {"class":"section menu_02"}).find_all("h3")
    # print(b4)
    for i in range(4):
        b3 = b2[i].get_text()
        food_list2.append(b3)

    df2 = pd.DataFrame(food_list1+food_list2, columns = ['menu'])
    df2.insert(0,'restaurant_name', '백철판')





    # 3번째 식당
    url = 'https://ssambap.co.kr/menu/' #원조쌈밥
    driver.get(url) 
    html = driver.page_source
    soup = bs(html, 'html.parser') 


    food_list3 = []
    e = soup.find_all("p", {"class":"post_tit"})


    for i in range(5):
        food_list3.append(e[i].get_text())

    df3 = pd.DataFrame(food_list3, columns = ['menu'])
    df3.insert(0,'restaurant_name', '백종원의 元祖쌈밥집')



    # 4번째 식당
    url = 'https://www.0410noodle.com/menu/' #미정국수
    driver.get(url) 
    html = driver.page_source
    soup = bs(html, 'html.parser') 

    food_list4 = []
    a = soup.find("section", {"class":"sub_sec menu_sec02 menu_sec noodle_menu"})
    b = a.find_all("div", {"class":"menu_cont"})

    for i in range(len(b)):
        food_list4.append(b[i].get_text().strip())

    c = soup.find("section", {"class":"sub_sec menu_sec03 menu_sec rice_menu"})
    d = c.find_all("div", {"class":"menu_cont"})
    for i in range(len(d)):
        food_list4.append(d[i].get_text().strip())

    df4 = pd.DataFrame(food_list4, columns = ['menu'])
    df4.insert(0,'restaurant_name', '미정국수0410')



    # 5번째 식당
    url = 'https://dolbaegi.com/#MENU' #돌배기
    driver.get(url) 
    html = driver.page_source
    soup = bs(html, 'html.parser') 

    food_list5=[]
    k = soup.find_all("div", {"class":"flip-box-back-inner"})
    for i in range(len(k)):
        j = k[i].h2.get_text()
        food_list5.append(j)

    df5 = pd.DataFrame(food_list5, columns = ['menu'])
    df5.insert(0,'restaurant_name', '돌배기')



    # 6번째 식당
    url = 'https://www.theborn.co.kr/theborn_brand/%eb%b3%b8%ea%b0%80/#menu_list' #본가
    driver.get(url) 
    html = driver.page_source
    soup = bs(html, 'html.parser')

    food_list6 = []
    m = soup.find_all("div", {"class":"name"})
    for i in range(len(m)):
        food_list6.append(m[i].get_text().replace("\n", ""))

    df6 = pd.DataFrame(food_list6, columns = ['menu'])
    df6.insert(0,'restaurant_name', '본가')


    dataframe_park = pd.concat([df1, df2, df3, df4, df5, df6])
    dataframe_park = dataframe_park.rename(columns={'menu':'menu_name'})
    dataframe_park.reset_index(drop = True, inplace = True)

    dataframe_final = dataframe_wonkkkn.append(dataframe_park)
    dataframe_final.reset_index(drop = True, inplace = True)

    # dataframe_final.to_csv("C:/crawring/menu_final.csv",index = False, encoding="utf-8-sig")








    return(dataframe_final)
```
