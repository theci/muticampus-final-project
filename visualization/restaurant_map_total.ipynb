```python
data = pd.read_excel("C:/crawring/restaurant_location_geo.xlsx")
data.head()
```

```python
data = data.loc[data["city"]=="서울특별시"]
```

```python
data.shape
```

```python
# # 위도
# latitude = 37.394946
# # 경도
# longitude = 127.111104
m = folium.Map(location=[37.559819, 126.963895],
               zoom_start=11, 
               width=750, 
               height=500
              )
m
```

```python
# 위도
latitude = data["Latitude"]
# 경도
longitude = data["Longitude"]
# 매장명
store_name = data["store"]
# 전화번호
telephone = data["telephone"]
# 도로명주소
address = data["address"]

for i,j,k,l,p in zip(latitude,longitude, store_name, telephone, address) :
    html = "<h3>{}<br><h5><b>지점연락처</b> : {}<br><b>도로명주소</b> : {}".format(k,l,p)
    iframe = folium.IFrame(html)
    popup = folium.Popup(iframe,
                     min_width=300,
                     max_width=500)
    
    folium.Marker([i,j],
                  popup= popup,
                  tooltip= k, 
                  icon=folium.Icon('red', icon='star'),
                 ).add_to(m)
m
```
